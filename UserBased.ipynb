```python
import pandas as pd

#user-items
def load_data():
    file='D:/作业/19.12课设/test.xlsx'
    data=pd.read_excel(open(file,'rb'),index=None,header=None,usecols=[0,1,2,3,4,5,6])
    data=data[:200]
    dict_data={}

    #将数据转成字典
    for user in range(7):
        for item in range(200):
            score = data[user][item]
            dict_data.setdefault(user,{})
            dict_data[user][item] = score

    return dict_data
```

```python
data=load_data()
```

```python
def item_users_average(data):
    item_users = dict()
    average_score = dict()
    for u,items in data.items():
        num = 0; sum_score = 0
        for i,score in items.items():
            item_users.setdefault(i,{})
            item_users[i].setdefault(u,0)
            item_users[i][u] = score

            if score != 0:
                num += 1
                sum_score += score

        average_score.setdefault(u,0)
        average_score[u] = sum_score / num
    return item_users,average_score
```

```python
item_users,average_score = item_users_average(data)
```

```python
from math import sqrt
def user_similarity(item_users,average_score):
    nu = dict()
    W = dict()
    for i, items in item_users.items():
        for u, rui in items.items():
            if rui == 0:  # 用户没有评分，直接跳过
                continue
            for v, rvi in items.items():
                if u == v or rvi == 0:  # 用户没有评分或者相同用户，直接跳过
                    continue  # u,v 必须同时对 i 产生评分

                nu.setdefault(u, {})
                nu[u].setdefault(v, 0)
                nu[u][v] += pow((rui - average_score[u]), 2)

                W.setdefault(u, {})
                W[u].setdefault(v, 0)
                W[u][v] += (rui - average_score[u]) * (rvi - average_score[v])

    for u in W:
        W[u] = {v: y / sqrt(nu[u][v] * nu[v][u]) for v, y in W[u].items()}

    return W
```

```python
def PredictAll(data, ave_vote, W, K=3):
    predict=dict()
    for user,items in data.items():
        for item in items:
            score=items[item]
            if score == 0:
                norm = 0
                for v, wuv in sorted(W[user].items(), key=itemgetter(1), reverse=True)[0:K]:  
                    # 和user相似度最高的K个用户
                    rvi = data[v][item]
                    score += wuv * (rvi - ave_vote[v])
                    norm += abs(wuv)
                if norm > 0:
                    score /= norm
                score += ave_vote[user]
                predict.setdefault(user,{})
                predict[user].setdefault(item,0)
                predict[user][item] = score

    return predict
```

```python
from operator import itemgetter
predict=PredictAll(data,average_score,W)
```
